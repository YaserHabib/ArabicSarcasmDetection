```python
import pandas as pd
```

```python
def siftData(data, name):
    data = data.sample(frac=1)
    data = data.drop_duplicates(subset=["tweet"])
    data = data.dropna()
    data = data.reset_index(drop=True)
    data.to_csv(rf"..\{name}.csv", index=False)

    return data
```

```python
augResult = pd.read_csv("augResult.csv")
backtransResult = pd.read_csv("backtransResult.csv")
synonymrepResult = pd.read_csv("synonymrepResult.csv")
originalDataset = pd.read_csv(r"https://raw.githubusercontent.com/iabufarha/ArSarcasm-v2/main/ArSarcasm-v2/training_data.csv")
GPTResult = pd.read_csv(r"https://raw.githubusercontent.com/YaserHabib/ArabicSarcasmDetection/main/Datasets/dataset_GPT.csv?token=GHSAT0AAAAAACHECZVVP45DRFZFCS7RBCBKZIZSEUQ")
```

```python
print(originalDataset.info(), "\n")
print("\n", augResult.info(), "\n")
print("\n", backtransResult.info(), "\n")
print("\n", synonymrepResult.info(), "\n")
print("\n", GPTResult.info(), "\n")
```

```python
augDataset = pd.concat([originalDataset, augResult])
GPTDataset = pd.concat([originalDataset, GPTResult])
fullDataset = pd.concat([originalDataset, augResult, GPTResult])
backtransDataset = pd.concat([originalDataset, backtransResult])
synonymrepDataset = pd.concat([originalDataset, synonymrepResult])
synGPTDataset = pd.concat([originalDataset, synonymrepResult, GPTResult])
backGPTDataset = pd.concat([originalDataset, backtransResult, GPTResult])
```

```python
print(originalDataset.info(), "\n")
print("\n", backtransDataset.info(), "\n")
print("\n", synonymrepDataset.info(), "\n")
print("\n", GPTDataset.info(), "\n")
print("\n", augDataset.info(), "\n")
print("\n", synGPTDataset.info(), "\n")
print("\n", backGPTDataset.info(), "\n")
print("\n", fullDataset.info(), "\n")
```

```python
GPTDataset = siftData(GPTDataset.copy(deep=True), "GPT Dataset")
fullDataset = siftData(fullDataset.copy(deep=True), "full Dataset")
augDataset = siftData(augDataset.copy(deep=True), "augmented Dataset")
synGPTDataset = siftData(synGPTDataset.copy(deep=True), "synGPT Dataset")
backGPTDataset = siftData(backGPTDataset.copy(deep=True), "backGPT Dataset")
synonymrepDataset = siftData(synonymrepDataset.copy(deep=True), "synrep Dataset")
backtransDataset = siftData(backtransDataset.copy(deep=True), "backtrans Dataset")
```
